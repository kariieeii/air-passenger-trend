```python
import pandas as pd
df = pd.read_csv('/content/AirPassengers.csv')
head = df.head()
tail = df.tail()

head, tail
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df['Month'], df['#Passengers'])
plt.title('Trend of Air Passengers Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.xticks(df['Month'][::12], rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
first_half = df['#Passengers'][:len(df)//2]
second_half = df['#Passengers'][len(df)//2:]
mean_first_half = first_half.mean()
variance_first_half = first_half.var()

mean_second_half = second_half.mean()
variance_second_half = second_half.var()

mean_first_half, variance_first_half, mean_second_half, variance_second_half
```

```python
threshold = 0.1

mean_change = abs(mean_first_half - mean_second_half)
variance_change = abs(variance_first_half - variance_second_half)

if mean_change < threshold and variance_change < threshold:
    print("The data appears to be stationary.")
else:
    print("The data appears to be non-stationary.")
```

```python
years = ['1972', '1973', '1974', '1975', '1976', '1977', '1978']
accidents = [2493, 2638, 2699, 3038, 3745, 4079, 4688]
plt.figure(figsize=(10, 6))
plt.bar(years, accidents, color='green', label='Number of Accidents')
plt.plot(years, accidents, marker='o', color='red', label='Trend Line')
plt.title('Number of Road Accidents in Punjab (1972-1978)')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
